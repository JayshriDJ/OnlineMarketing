```python
df=pd.read_csv("online.csv")
df
```

```python
df.rename({"Marketing Spend":"Marketing_Spend"},axis=1,inplace=True)
```

```python
df["Area"].replace({'Dhaka': 0, 'Ctg': 1, 'Rangpur': 2},inplace=True)
```

```python
x=df.drop("Profit",axis=1)
y=df["Profit"]
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=30)
```

```python
lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
```

```python
yp_train=lr_model.predict(x_train)
mse=mean_absolute_error(y_train,yp_train)
print("MSE",mse)

r2=r2_score(y_train,yp_train)
print("R2 score:",r2)


```

```python
yp_test=lr_model.predict(x_test)
mse=mean_absolute_error(y_test,yp_test)
print("MSE",mse)

r2=r2_score(y_test,yp_test)
print("R2 score:",r2)
```
